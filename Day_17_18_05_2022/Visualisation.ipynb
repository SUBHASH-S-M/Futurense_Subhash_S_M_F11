```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cordinates_y=np.array([3,4,6,7,9,15,16])     #x axis will be geenrated
plt.plot(cordinates_y)
plt.show()
```

```python
plt.plot(cordinates_x, cordinates_y, 'go--', linewidth=2, markersize=10)
```

```python
plt.plot(cordinates_x, cordinates_y, marker='o')
#if marker='o' is specified then it drwas a line and connects the points
```

```python
plt.plot(cordinates_x, cordinates_y,'o')
#if marker='o' is not specified then it just points
```

```python
plt.plot(cordinates_x, cordinates_y, marker='o', linestyle='dashed',
     linewidth=2, markersize=12)
```

```python
import seaborn as sns
def sinplot(flip=1):
    x=np.linspace(0,14,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
```

```python
sns.set_style("white")
sns.set_palette("flare")
sinplot()
plt.show()
```
